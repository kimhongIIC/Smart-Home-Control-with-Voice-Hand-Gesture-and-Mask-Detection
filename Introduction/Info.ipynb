# Smart Home 

### There are two important things in order to run this project:
- Arduino code
- Python code

### Arduino code   
- Include [Firmata](https://www.arduino.cc/reference/en/libraries/firmata/) libraries to your Arduino IDE 
- Upload StandardFirmata to Arduino uno but you have to change Baud rate to 9600 please watch this [video](https://youtu.be/L3wjZOAyxEE?si=93w8De377jUCl73o)


### Python code
- You have to install Python version 3.8 (the best version for this project) watch this [video](https://youtu.be/3hzV1_FwWu0?si=WJha5i_4jEwEMEu4)
- Make sure that you can use 'pip' to inistall python packages. If you have any errors with 'pip' please wach some tutorial on youtube to solve it by yourself.
- Open terminal and install file requirement(this file include packages that need to run this project) by typing: pip install -r requirements.txt
- Don't forget to connect bluetooth with smart home.
- If you have done all step above, it time to run the code. Please run home.py

# NOTE: 
- Be careful with speech recognition cuz you need to connect to internet in order to use this function.
- If it still has any errors, please try your best to solve them. Nothing is easy if you don't try it by yourself.
